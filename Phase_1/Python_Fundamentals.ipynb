What does **Break** and **Continue** do in while loops?

<img src= 'https://files.realpython.com/media/t.899f357dd948.png' width = 300>

```python
# example of break 
i = 1
while i < 10:
    i += 1 
    if i == 3:
        break
    print(i)
```

```python
# example of continue
i = 1
while i < 10:
    i += 1
    if i == 3:
        continue
    print(i)
```

```python
# nested while loops -- OPTIONAL 
i = 1
j = 5
while i < 10:
    while j < 10: 
        print("now:", i,j)
        i += 1
        j += 1
        print(i,j)        
        
    i += 1
```
